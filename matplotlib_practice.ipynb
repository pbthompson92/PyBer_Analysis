```python
import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of strings for each month 
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# set the y-axis to a list of floats as the total fare in US dollars accululated for each month 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax= plt.axes()
ax.plot(x_axis,y_axis)
plt.show()
```

```python
# create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston', color = 'blue', linewidth = 2, marker = '*')

# create labels for the x and y axis
plt.xlabel('Date')
plt.ylabel('Fare($)')

#set y limit between 0 to 45
plt.ylim(0,45)

#create a title
plt.title('PyBer Fare by Month')

# add the legend
plt.legend()

#add a grid
plt.grid()

plt.subplots(figsize=(20, 10))
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis, y_axis, color = "green", marker = 'd', lw = 2, label = 'Boston')
ax.set_xlabel('Date')
ax.set_ylabel('Fares($)')
ax.set_ylim(0,45)
ax.grid()
ax.legend()
ax.set_title('PyBer Fare by Month')
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c='r', label = 'Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s= y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s= y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i*5 for i in y_axis], label = 'Boston', c='skyblue', edgecolor = 'black', lw = 2, alpha = .8)
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend()
```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
plt.subplots(figsize = (8,8))
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, 
        explode = explode_values, 
        labels=x_axis, 
        autopct='%.1f%%',
        colors = colors)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,8))
explode_values_2 = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,
       labels=x_axis,
       explode = explode_values_2,
       colors = colors,
       autopct = '%.1f%%',
       startangle = 90,
       counterclock = False,
       shadow = True)
plt.show()
```
